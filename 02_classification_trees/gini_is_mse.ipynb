```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
p = np.arange(0, 1, 0.01)
q = (1-p)
entropy = - (p * np.log2(p + 0.001) + q * np.log2(q + 0.001))

gini = p*q + q*p

misclassified = np.where(p < q, p, q)

MSE = p*q*q + q*p*p

plt.plot(p, entropy/entropy.max(), label='entropy')
plt.plot(p, gini/gini.max(), label='gini')
plt.plot(p, misclassified/misclassified.max(), label='misclassified')
plt.plot(p, MSE/MSE.max(), label='MSE')

plt.legend()
plt.xlabel('p')
plt.ylabel('error')
plt.show()
```
