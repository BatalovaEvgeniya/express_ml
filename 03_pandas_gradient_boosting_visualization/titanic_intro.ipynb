```python
training_columns = ['PassengerId', 'Pclass', 'Age', 'SibSp', 
                    'Parch', 'Fare', 'Gender', 'Embarked'] 
X = data[training_columns].fillna(0) 
y = data['Survived']

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, 
                             oob_score=True,
                             random_state=1)
```

```python
clf.fit(X, y) 
```

```python
clf.predict(X) 
```

```python
clf.oob_score_
```

```python
clf.feature_importances_
```

```python
sorted(zip(clf.feature_importances_, training_columns), 
       reverse=True)
```
